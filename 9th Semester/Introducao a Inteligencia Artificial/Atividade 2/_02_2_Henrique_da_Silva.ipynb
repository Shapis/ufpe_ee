```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import librosa
import librosa.display

from IPython.display import Audio
```

```python
fname = 'camoes.wav'
y, sr = librosa.load(fname, sr=None, mono=True)
```

```python
Audio(y, rate=sr)
```

```python
S = librosa.stft(y, n_fft=2048, hop_length=512, win_length=1024)
S = np.abs(S)
librosa.display.specshow(librosa.amplitude_to_db(S, ref=np.max), y_axis='log', x_axis='time', sr=sr)
```
